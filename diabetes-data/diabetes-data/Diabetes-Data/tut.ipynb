```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_fwf("data-01")
```

```python
df
```

```python
df[['Date','Time', 'Category', 'Target']] = df['Date\tTime\tCategory\tTarget'].str.split("\t", expand=True)
```

```python
df
```

```python
df.drop('9:09\t58\t100', axis = 1, inplace=True)
```

```python
df.rename({'04-21-1991': 'Date'}, axis=1, inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Time'] = pd.to_datetime(df['Time'], errors='coerce')
```

```python
df
```

```python
df['Time'] = df['Time'].dt.time
```

```python
df.sort_values(['Date', 'Time'], ascending=[True, True], inplace=True)
```

```python
df.to_csv("result.csv", index=False)
```
